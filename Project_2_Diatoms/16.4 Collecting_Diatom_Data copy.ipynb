```python
import os

# Set the path to your image folder
diatom_images = 'Diatoms/images'
annotations = 'Diatoms/annotations'
outlines_image = 'diatomsORG_bw_outlines'


# Define common image extensions
image_extensions = ('.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff')

annotations_image_count = sum(
    1 for file in os.listdir(annotations)
    if file.lower().endswith(image_extensions)
)
# Must download and insert the image file into Diatoms/ folder locally

# diatom_image_count = sum(
#     1 for file in os.listdir(diatom_images)
#     if file.lower().endswith(image_extensions)
# )

annotations_image_count = sum(
    1 for file in os.listdir(annotations)
    if file.lower().endswith(image_extensions)
)
import os

# List all batch folders
batch_folders = [f'diatomsORG/batch{i}' for i in range(1, 11)]
batch_bw_folders = [f'diatomsORG_bw_outlines/batch{i}' for i in range(1, 6)]

diatom_image_count = sum(
    1 for file in os.listdir(diatom_images)
    if file.lower().endswith(image_extensions)
)

annotations_image_count = sum(
    1 for file in os.listdir(annotations)
    if file.lower().endswith(image_extensions)
)

# Define image extensions to look for
image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.tif', '.tiff', '.gif')

total_web_images = 0

for folder in batch_folders:
    if os.path.exists(folder):
        images_in_folder = [
            f for f in os.listdir(folder)
            if f.lower().endswith(image_extensions)
        ]
        print(f"{folder}: {len(images_in_folder)} images")
        total_web_images += len(images_in_folder)
    else:
        print(f"{folder}: Folder not found!")

webbw_image_count = sum(
    1 for file in os.listdir(outlines_image)
    
    if file.lower().endswith(image_extensions)
)

print(f"\nTotal images across all batches: {total_web_bw_images}")
print(f"\nTotal images across all batches: {total_web_images}")


print("Total: ", total_web_images+webbw_image_count+ annotations_image_count + diatom_image_count)



# Will not run unless image file is inserted into the Diatoms folder
# print(f"Number of image files: {diatom_image_count + annotations_image_count+ web_image_count + webbw_image_count}")

```
